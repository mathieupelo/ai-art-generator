```python
from mtgsdk import Card

# Get all cards
#cards = Card.all()

# Filter Cards
# You can chain 'where' clauses together. The key of the hash
# should be the URL parameter you are trying to filter on
cards = Card.where(setName='Khans of Tarkir') \
            .where(type='creature') \
            .all()
            

# Get cards on a specific page / pageSize
#cards = Card.where(page=50).where(pageSize=50).all()
```

```python
print(cards)
```

```python
for c in cards:
    print(c.name, c.image_url)
```

```python
import pandas as pd

# Create a list of dictionaries containing the name and image_url of each card
card_data = [{'Name': c.name, 'Image_URL': c.image_url} for c in cards]

# Create a DataFrame from the card_data list
df = pd.DataFrame(card_data)

# Print the DataFrame
print(df)
```

```python
df.to_csv('dataset/TKT.csv', index=False)
```

```python
import pandas as pd

df = pd.read_csv('dataset\\raw\\TKT.csv')

# Print the DataFrame
print(df)
```

```python
import os
import requests

# Create the folder if it doesn't exist
os.makedirs('dataset/images', exist_ok=True)

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    name = row['Name']
    image_url = row['Image_URL']
    
    # Skip rows with missing image URLs
    if pd.isnull(image_url):
        continue
    
    # Download the image
    response = requests.get(image_url)
    
    # Save the image with the name in jpg format
    with open(f'dataset/images/{name}.PNG', 'wb') as f:
        f.write(response.content)
```
